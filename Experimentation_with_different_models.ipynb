# 1. Face Detection

For the FER, it it better if the face is cropped out from the picture/image. For that a face detection model will be implemented.

## 1.1. Haar Cascades with OpenCV

```python
import cv2
# Load the pre-trained Haar Cascade classifier
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
import matplotlib.pyplot as plt
image = cv2.imread("MSc-Project-Expression-Recognition-/man.jpg")
plt.imshow(image)
plt.axis(False)
plt.show();
```

```python
from google.colab.patches import cv2_imshow as image_show
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Perform face detection
faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# Draw rectangles around detected faces
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)

# Display the output

plt.imshow(image)
plt.axis(False)
plt.show();
```
